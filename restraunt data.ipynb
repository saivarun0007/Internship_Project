## Importing LIBRARIES

## Dropping the DUPLICATES

```python
df.drop_duplicates(inplace=True)
df.shape
```

## Cleaning the RATE COLUMN

```python
df['rate'].unique()
```

## REMOVING "NEW"; "-"; "/5" from RATE COLUMN

```python
def handlerate(value):
    if(value=='NEW' or value =='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
df['rate'].tail() 
```

## Filling NULL VALUES in RATE COLUMN

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isnull().sum()
df.info()
```

## DROPPING NULL VALUES

```python
df.dropna(inplace=True)
df.head()
```

```python
df.tail()
```

```python
df.rename(columns={'approx_cost(for two people)':'costtwoplates', 'listed_in(type)':'type'},inplace=True)
df.head()
```

```python
df.tail()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

## Here we have LISTED IN(CITY) & LOCATION  from this keep only ONE

```python
df=df.drop(['listed_in(city)'],axis=1)
```

```python
df['costtwoplates'].unique()
```

## MAKING "COST TWO PLATES" as UNIQUE

```python
def handle_comma(value):
    value= str(value)
    if ',' in value:
        value= value.replace(',', '')
        return float(value)
    else:
        return float(value)
df['costtwoplates']= df['costtwoplates'].apply(handle_comma)
df['costtwoplates'].unique()
```

```python
df.head()
```

```python
df.tail()
```

## Cleaning REST TYPE COLUMN

```python
rest_types=df['rest_type'].value_counts(ascending=False)
rest_types
```

```python
rest_types_lessthan2000=rest_types[rest_types<2000]
rest_types_lessthan2000
```

## Making "REST TYPES LESSTHAN 2000" as Others

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan2000):
        return 'others'
    else:
        return value
df['rest_type']=df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

## Cleaning "LOCATION COLUMN"

```python
location=df['location'].value_counts(ascending=False)
location_lessthan300= location[location<300]
def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
df['location']= df['location'].apply(handle_location)
df['location'].value_counts()
```

## Cleaning "CUISINES COLUMN"

```python
cuisines= df['cuisines'].value_counts(ascending=False)
cuisines_lessthan200= cuisines[cuisines<200]
def handle_cuisines(value):
    if (value in cuisines_lessthan200):
        return'others'
    else:
        return value
df['cuisines']= df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

```python
df.head()
```

```python
df.tail()
```
